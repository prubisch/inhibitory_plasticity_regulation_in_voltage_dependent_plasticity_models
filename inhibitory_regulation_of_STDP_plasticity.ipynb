```python
from brian2 import *
from brian2tools import *

import matplotlib.pyplot as plt
%matplotlib inline


from plotting_helper import * 


figures_path = 'figures/'
palette = plt.rcParams['axes.prop_cycle'].by_key()['color']  


def LTP_LTD_ratio(kernel):
    
    if np.any(kernel == np.nan) or np.all(kernel == 0): 
        return np.nan
    
    total_int = np.sum(np.abs(kernel))
    
    LTP_mask = 1.0*(kernel > 0)
    LTD_mask = 1.0*(kernel<0)
    
    LTP_perc = np.sum(kernel*LTP_mask)/total_int
    LTD_perc = np.sum(kernel*LTD_mask)/total_int #negative value anyway
    
    return LTP_perc+LTD_perc   

```

```python
%colors nocolor


# VDP parameters
##########################################
c_ca = 3.0
act_offset = 0
tau_Ca = 1 * ms
tau_w = 1 * ms

#
# VDP equations
##########################################

VDP_eq = '''
    dglutamate/dt = - glutamate/tau_glutamate : 1 (clock-driven)
    sig_LTP = 1/(1+exp(k_LTP*(-(v - theta_NMDA)+0*mV))) : 1 (constant over dt)
    dv_NMDA_rise/dt = (-v_NMDA_rise + clip(clip(sig_LTP - v_NMDA,0,100) - v_NMDA_rise,0,100))/tau_NMDA_rise : 1 (clock-driven)             
    dv_NMDA/dt = (NMDA_amp * v_NMDA_rise - v_NMDA)/tau_NMDA_fall : 1 (clock-driven)
    sig_LTD = 1/(1+exp(k_LTD*(-(v - theta_VDCC)+0*mV))) : 1 (constant over dt)
    dVDCC_rise/dt = (-VDCC_rise + clip(clip(sig_LTD - VDCC,0,100) -VDCC_rise, 0, 100))/tau_VDCC_rise : 1 (clock-driven) 
    dVDCC/dt = (VDCC_amp * VDCC_rise - VDCC) * 1/tau_VDCC_fall : 1 (clock-driven)
    Ca = VDCC**(exp(-c_ca)*20) + v_NMDA**(exp(-c_ca)*20) * int(glutamate>0) : 1 (constant over dt)
    dw/dt = clip(eta * Ca  * (v_NMDA * LTP_amp - VDCC * LTD_amp) * glutamate/tau_w*dt,(wmin-w),(wmax-w))/dt : 1 (clock-driven)
    tau_glutamate :second
    theta_NMDA : volt
    tau_NMDA_rise : second
    tau_NMDA_fall : second
    NMDA_amp : 1
    theta_VDCC : volt
    tau_VDCC_rise : second
    tau_VDCC_fall : second
    VDCC_amp : 1
    k_LTP : 1/volt
    k_LTD : 1/volt
    LTP_amp : 1
    LTD_amp : 1
    Apre : 1  #this is technically decided by tau_glutamate but just adding it here for completeness sake
    eta : 1
    c_ca : 1
    wmin :1
    wmax:1
'''

eonpre_VDP = '''
            glutamate += Apre'''
eonpre_VDP_active = '''
            glutamate += Apre
            g_syn_e += IG_g_e'''

clopath_eq = '''
    dvp/dt = (-vp + v)/tau_p : volt (clock-driven)
    dvm/dt = (-vm + v)/tau_m : volt (clock-driven)
    ds/dt = -s/tau_s : 1 (clock-driven)
    dw/dt = eta * A_p  * (v - theta_p)/mV * int((v - theta_p)/mV > 0 ) * (vp - theta_m)/mV*int((vp - theta_m)/mV > 0 ) * s / tau_w : 1 (clock-driven)
    tau_p : second
    tau_m : second
    tau_s : second
    A_p : 1
    A_d : 1
    theta_p : volt
    theta_m : volt
    g : 1
    eta : 1

'''

eonpre_clopath ='''s += g
                w -= eta * 0.1 * A_d *(vm - theta_m)/mV * int((vm - theta_m)/mV > 0)'''


eonpre_clopath_active ='''s += g
                        w -= eta * 0.1 * A_d *(vm - theta_m)/mV * int((vm - theta_m)/mV > 0)
                        g_syn_e += IG_g_e'''


meissner_bernard_eq = '''
    dvp/dt = (-vp + v)/tau_p : volt (clock-driven)
    dvm/dt = (-vm + v)/tau_m : volt (clock-driven)
    ds/dt = -s/tau_s : 1 (clock-driven)
    dw/dt = eta * (A_p  * (vp - theta_p)/mV*int((vp - theta_p)/mV > 0 ) - A_d *(vm - theta_m)/mV * int((vm - theta_m)/mV > 0)) * s / tau_w : 1 (clock-driven)
    tau_p : second
    tau_m : second
    tau_s : second
    A_p : 1
    A_d : 1
    theta_p : volt
    theta_m : volt
    g : 1
    eta : 1

'''

eonpre_mb = '''s += g'''
eonpre_mb_active = '''s += g
                    g_syn_e += IG_g_e'''

#same for all
ionpre = '''
            g_syn_i += IG_g_i
            '''

GLIF_eqs = '''
    dv/dt = (1/C) * (I_stim(t, i) + I_1 + I_2 - (v - E_L)/R_ASC - g_syn_i * (v-E_inh)/R_ASC - g_syn_e * (v - E_ex) / R_ASC) : volt 
    dI_1/dt = -k_1 * I_1 : amp 
    dI_2/dt = -k_2 * I_2 : amp 
    dTheta_s/dt = -1./one_over_b_spike * Theta_s : volt 
    dTheta_v/dt = av * (v - E_L) - bv * Theta_v : volt 
    Theta_total = Theta_inf + Theta_s + Theta_v : volt
    dg_syn_i/dt = -g_syn_i/tau_syn_i : 1
    dg_syn_e/dt = -g_syn_e/tau_syn_i : 1
    C: farad
    E_L: volt
    R_ASC: ohm
    k_1: Hz
    k_2: Hz
    one_over_b_spike: second
    av: Hz
    bv: Hz
    Theta_inf: volt
    f_v: 1
    delta_v: volt
    delta_Theta_inf: volt
    delta_t: second
    delta_I_1: amp
    delta_I_2: amp
    
'''
GLIF_reset = '''
    v = E_L + f_v * (v - E_L) + delta_v
    Theta_s = Theta_s + delta_Theta_inf
    f_1 = exp(-k_1 * delta_t)
    f_2 = exp(-k_2 * delta_t)
    I_1 = f_1 * I_1 + delta_I_1
    I_2 = f_2 * I_2 + delta_I_2
    Theta_v = Theta_v
'''


######################################################################
### functions for setting up the neuron model and plasticity models
######################################################################

def unpack_params_VDP(Syn, params): 
    
    Syn.Apre = params[0]
    Syn.tau_glutamate = params[1] *ms
    
    Syn.k_LTD = params[2] /mV 
    LTD_rise = params[3] 
    LTD_fall = params[4] 
    
    Syn.tau_VDCC_rise = LTD_rise * ms
    Syn.tau_VDCC_fall = LTD_fall * ms
    Syn.theta_VDCC  = params[5] * mV 
    Syn.LTD_amp = params[6]
    
    Syn.k_LTP = params[7] / mV
    LTP_rise = params[8] 
    LTP_fall = params[9] 
    Syn.tau_NMDA_rise = LTP_rise * ms
    Syn.tau_NMDA_fall =  LTP_fall * ms
    Syn.theta_NMDA = params[10] * mV 
    Syn.LTP_amp =  params[11]
    Syn.eta = params[12]
    Syn.wmin = params[13]
    Syn.wmax = params[14]
    Syn.c_ca = params[15]
    
    #predetermined by time constants
    Syn.NMDA_amp = (LTP_rise/LTP_fall)**(LTP_fall/(LTP_rise - LTP_fall))
    Syn.VDCC_amp = (LTD_rise/LTD_fall)**(LTD_fall/(LTD_rise - LTD_fall))
    
    
    return Syn


def unpack_params_clopath(Syn, params):
    
    Syn.g = params[0] 
    Syn.tau_s = params[1] *ms
    
    Syn.tau_m = params[2] *ms
    Syn.tau_p = params[3] * ms
    Syn.A_d = params[4]
    Syn.A_p = params[5]
    Syn.theta_m = params[6] *mV
    Syn.theta_p = params[7] * mV
    Syn.eta = params[8]
    
    return Syn


import pandas as pd
global df
if 'df' not in locals():
    df = pd.read_pickle('glif_params.pkl')
    df = df.drop(['13'])
    df = df.set_index([pd.Index(np.arange(15))])
    
def unpack_GLIF5_params(neuron_pop, cluster_id): 
    
    tau = df.iloc[cluster_id].at['$\tau$ (ms)'] # ms membrane constant
    #tau = tau * 1e-3 # S
    
    neuron_pop.C = df.iloc[cluster_id].at['C (pF)']*pF
    neuron_pop.E_L = df.iloc[cluster_id].at['El (mV)'] *mV
    neuron_pop.Theta_inf = df.iloc[cluster_id].at['th$_\\infty$ (mV)'] *mV
    neuron_pop.delta_Theta_inf = df.iloc[cluster_id].at['$\Delta$ th$_\\infty$ (mV)'] *mV
    neuron_pop.R_ASC = df.iloc[cluster_id].at['$R_{ASC} (M\\Omega$)'] *Mohm
    neuron_pop.delta_t = df.iloc[cluster_id].at['$\delta t$ (ms)'] *ms
    neuron_pop.f_v = df.iloc[cluster_id].at['slope'] #1
    neuron_pop.delta_v = df.iloc[cluster_id].at['intercept (mV)'] *mV
    neuron_pop.one_over_b_spike = df.iloc[cluster_id].at['1/($b_{spike}$) (s)'] *second
    neuron_pop.k_1 = df.iloc[cluster_id].at['$k_1$ (Hz)'] *Hz
    neuron_pop.k_2 = df.iloc[cluster_id].at['$k_2$ (Hz)'] *Hz
    neuron_pop.delta_I_1 = df.iloc[cluster_id].at['$\delta I_1$ (pA)'] *pamp
    neuron_pop.delta_I_2 = df.iloc[cluster_id].at['$\delta I_2$ (pA)'] *pamp
    neuron_pop.av = df.iloc[cluster_id].at['$a_v$'] *Hz
    neuron_pop.bv = df.iloc[cluster_id].at['$b_v$'] *Hz
    
    #noise constant 
    tau_n = 2*tau
        
    return neuron_pop

###########################
# simulation protocol
###########################

def inhib_STDP_sim(stdp_dts, t_inhib, inhibition_strength, synapse_type):
    #stdp dts for kernel
    #t_inhib relative to postspikes
    #params  for synapses
    start_scope()
    N = int(stdp_dts.shape[0])
    w0 = 1

    GE_ = NeuronGroup(N, GLIF_eqs, threshold='v > Theta_total',  method='euler', reset=GLIF_reset) #GLIF ohne refraktaer zeit 
                      #refractory = df.iloc[neuron_cluster].at['$\delta t$ (ms)'] *ms)
    GE = unpack_GLIF5_params(GE_ , neuron_cluster)
    GE.v = GE.E_L
    
    I_spikes = np.zeros(N)
    pre_time = 200
    pre_spikes = np.zeros(N, dtype = 'int')
    for post_ind, post in enumerate(stdp_dts): 
        pre_spikes[post_ind] = pre_time
        I_spikes[post_ind] = pre_time + post+t_inhib
        
    IG_E = SpikeGeneratorGroup(N, np.arange(N), pre_spikes*ms)
    IG_I = SpikeGeneratorGroup(N, np.arange(N), I_spikes*ms)



    # Synapses
    ##########################################

    S_IG_I = Synapses(IG_I, GE,
                on_pre=ionpre, 
                method='euler')
    S_IG_I.connect(condition='i == j')
    
    
    if synapse_type == 'VDP': 
    
        S_IG_E = Synapses(IG_E, GE, VDP_eq,
                    on_pre=eonpre_VDP,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_VDP(S_IG_E, vdp_params)
        p_mon = StateMonitor(S_IG_E, ['v_NMDA', 'VDCC'], record = 1, dt = .1*ms)
    
    elif synapse_type == 'Clopath': 
        S_IG_E = Synapses(IG_E, GE, clopath_eq,
                    on_pre=eonpre_clopath,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_clopath(S_IG_E, clopath_params)
        S_IG_E.vm = GE.E_L
        S_IG_E.vp = GE.E_L
        
    elif synapse_type == 'Meissner_Bernard': 
        S_IG_E = Synapses(IG_E, GE, meissner_bernard_eq,
                    on_pre=eonpre_mb,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_clopath(S_IG_E, mb_params)
        S_IG_E.vm = GE.E_L
        S_IG_E.vp = GE.E_L
        
    else: 
        print('Plasticity model not recognised!')
        raise Exception('Plasticity model not recognised!')


    # Monitors
    ##########################################

    M_IG_E_w = StateMonitor(S_IG_E, ['w'], record=np.arange(N), dt=1 * ms)
    spikes = SpikeMonitor(GE)
    v_mon = StateMonitor(GE, ['v'], record = 1, dt = .01*ms)
    ##########################################

    I_stim_duration = 8#0.6ms 
    I_stim_amp = 5 * nA #500*mV this is for LIF neurons

    IG_g_i = inhibition_strength
    IG_g_e = 1

    post_stims = np.zeros((10000, N))*amp #*volt
    for ind, stdp_dt in enumerate(stdp_dts):
        post_stims[int((pre_time+stdp_dt)/0.1):int((pre_time+stdp_dt)/0.1)+I_stim_duration, ind] = I_stim_amp
    I_stim = TimedArray(post_stims, dt=.1*ms)
    
    net = Network()
    if synapse_type == 'VDP':
        net.add(GE, IG_E, IG_I, S_IG_E, S_IG_I, M_IG_E_w,spikes, v_mon, p_mon)
    else: 
        net.add(GE, IG_E, IG_I, S_IG_E, S_IG_I, M_IG_E_w,spikes, v_mon)

    S_IG_E.w = w0

    net.run(1000*ms)
    perc_w = M_IG_E_w.w[:,-1]-w0
    
    return perc_w



def inhib_pre_STDP_sim(stdp_dts, t_inhib, inhibition_strength, synapse_type):
    #stdp dts for kernel
    #t_inhib relative to prespikes
    #params  for synapses
    start_scope()
    N = int(stdp_dts.shape[0])
    w0 = 1

    GE_ = NeuronGroup(N, GLIF_eqs, threshold='v > Theta_total',  method='euler', reset=GLIF_reset)#, refractory = df.iloc[neuron_cluster].at['$\delta t$ (ms)'] *ms)
    GE = unpack_GLIF5_params(GE_ , neuron_cluster)
    GE.v = GE.E_L
    
    I_spikes = np.zeros(N)
    pre_time = 200
    pre_spikes = np.zeros(N, dtype = 'int')
    for post_ind, post in enumerate(stdp_dts): 
        pre_spikes[post_ind] = pre_time
        I_spikes[post_ind] = pre_time +t_inhib
        
    IG_E = SpikeGeneratorGroup(N, np.arange(N), pre_spikes*ms)
    IG_I = SpikeGeneratorGroup(N, np.arange(N), I_spikes*ms)



    # Synapses
    ##########################################

    S_IG_I = Synapses(IG_I, GE,
                on_pre=ionpre, 
                method='euler')
    S_IG_I.connect(condition='i == j')
    
    
    if synapse_type == 'VDP': 
    
        S_IG_E = Synapses(IG_E, GE, VDP_eq,
                    on_pre=eonpre_VDP,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_VDP(S_IG_E, vdp_params)
    
    elif synapse_type == 'Clopath': 
        S_IG_E = Synapses(IG_E, GE, clopath_eq,
                    on_pre=eonpre_clopath,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_clopath(S_IG_E, clopath_params)
        
    elif synapse_type == 'Meissner_Bernard': 
        S_IG_E = Synapses(IG_E, GE, meissner_bernard_eq,
                    on_pre=eonpre_mb,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_clopath(S_IG_E, mb_params)
        S_IG_E.vm = GE.E_L
        S_IG_E.vp = GE.E_L
        
    else: 
        raise Exception('Plasticity model not recognised!')

    # Monitors
    ##########################################

    M_IG_E_w = StateMonitor(S_IG_E, ['w'], record=np.arange(N), dt=1 * ms)
    ##########################################

    I_stim_duration = 8 #0.6ms 
    I_stim_amp = 5 * nA #500*mV this is for LIF neurons

    IG_g_i = inhibition_strength
    IG_g_e = 1

    post_stims = np.zeros((10000, N))*amp
    for ind, stdp_dt in enumerate(stdp_dts):
        post_stims[int((pre_time+stdp_dt)/0.1):int((pre_time+stdp_dt)/0.1)+I_stim_duration, ind] = I_stim_amp
    I_stim = TimedArray(post_stims, dt=.1*ms)
    
    net = Network()
    net.add(GE, IG_E, IG_I, S_IG_E, S_IG_I, M_IG_E_w)

    S_IG_E.w = w0

    net.run(1000*ms)
    perc_w = M_IG_E_w.w[:,-1]-w0
    
    return perc_w


```

```python
def inhib_STDP_sim_stim(stdp_dts, dt_i, i_length, inhibition_strength, exc_strength, E_L, synapse_type):
    #stdp dts for kernel
    #t_inhib relative to postspikes
    #params  for synapses
    start_scope()
    N = int(stdp_dts.shape[0])
    w0 = 1

    GE = NeuronGroup(N, 'v = I_stim(t,i) : volt ', threshold='v > -50 * mV',  method='euler', refractory = 4 * ms) #GLIF ohne refraktaer zeit 
                   #refractory = df.iloc[neuron_cluster].at['$\delta t$ (ms)'] *ms)
    
    #GE.v = E_L * mV
    
    pre_time = 200
    pre_spikes = np.zeros(N, dtype = 'int')
    for post_ind, post in enumerate(stdp_dts): 
        pre_spikes[post_ind] = pre_time
        
    IG_E = SpikeGeneratorGroup(N, np.arange(N), pre_spikes*ms)



    # Synapses
    ##########################################
    
    
    if synapse_type == 'VDP': 
    
        S_IG_E = Synapses(IG_E, GE, VDP_eq,
                    on_pre=eonpre_VDP,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_VDP(S_IG_E, vdp_params)
        p_mon = StateMonitor(S_IG_E, ['v_NMDA', 'VDCC'], record = 1, dt = .1*ms)
    
    elif synapse_type == 'Clopath': 
        S_IG_E = Synapses(IG_E, GE, clopath_eq,
                    on_pre=eonpre_clopath,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_clopath(S_IG_E, clopath_params)
        S_IG_E.vm = E_L* mV
        S_IG_E.vp = E_L* mV
        
        
    elif synapse_type == 'Meissner_Bernard': 
        S_IG_E = Synapses(IG_E, GE, meissner_bernard_eq,
                    on_pre=eonpre_clopath,
                    method='euler')
        S_IG_E.connect(condition='i == j')
        S_IG_E = unpack_params_clopath(S_IG_E, mb_params)
        S_IG_E.vm = E_L* mV
        S_IG_E.vp = E_L* mV
        
    else: 
        print('Plasticity model not recognised!')
        raise Exception('Plasticity model not recognised!')


    # Monitors
    ##########################################

    M_IG_E_w = StateMonitor(S_IG_E, ['w'], record=np.arange(N), dt=1 * ms)
    spikes = SpikeMonitor(GE)
    v_mon = StateMonitor(GE, ['v'], record = 1, dt = .01*ms)
    ##########################################

    I_stim_duration = 40 # 2.5ms #8#0.6ms 

    IG_g_i = inhibition_strength
    IG_g_e = 1

    post_stims = np.zeros((10000, N)) * volt
    post_stims[:,:] = E_L* mV
    for ind, stdp_dt in enumerate(stdp_dts):
        post_stims[int((pre_time+stdp_dt)/0.1):int((pre_time+stdp_dt)/0.1)+I_stim_duration, ind] = post_stims[int((pre_time+stdp_dt)/0.1):int((pre_time+stdp_dt)/0.1)+I_stim_duration, ind] + exc_strength * mV
        post_stims[int((pre_time+stdp_dt+dt_i-i_length)/0.1):int((pre_time+stdp_dt+dt_i)/0.1), ind] =post_stims[int((pre_time+stdp_dt+dt_i-i_length)/0.1):int((pre_time+stdp_dt+dt_i)/0.1), ind] - inhibition_strength * mV
    I_stim = TimedArray(post_stims, dt=.1*ms)
    
    net = Network()
    if synapse_type == 'VDP':
        net.add(GE, IG_E,  S_IG_E,  M_IG_E_w,spikes, v_mon, p_mon)
    else: 
        net.add(GE, IG_E,  S_IG_E, M_IG_E_w,spikes, v_mon)

    S_IG_E.w = w0

    net.run(1000*ms)
    perc_w = M_IG_E_w.w[:,-1]-w0
    
    
    return perc_w, np.max(v_mon.v/mV)

```

```python
###### compare there different models
neuron_cluster = 0
E_inh = -80 * mV
E_ex = 0 * mV
stdp_delays = np.arange(-50,50,0.5)
tau_syn_i = 2 * ms
inhibition_strengths = np.array([5])#np.arange(0,105, 5)#np.array([0, 60,70,80,90])
t_i = 0
i_lengths = np.arange(5,25,1)
repetitions = 100
E_L = -79
e_strength = 30


#define plasticity parameters
vdp_params = [0.1, 10, 0.4, 1,20, -43.75,0.18,0.4, 1, 5, -40.75,1.0,1,0,2.0,3.0]
clopath_params = [1/15, 15, 10,7, 1.e-2, 2.e-4,-75, -68,.1]
mb_params = [1/15, 15, 10,7, 2.e-4, 2.e-3,-75, -70,10]


ref_lenvdp = LTP_LTD_ratio(inhib_STDP_sim_stim(stdp_delays, 0,5, 5, e_strength, E_L,'VDP')[0])
ref_lencl = LTP_LTD_ratio(inhib_STDP_sim_stim(stdp_delays, 0,5, 5,e_strength,  E_L,'Clopath')[0])
ref_lenmb = LTP_LTD_ratio(inhib_STDP_sim_stim(stdp_delays, 0,5, 5, e_strength, E_L,'Meissner_Bernard')[0])


lenvdp_ratios = np.zeros([inhibition_strengths.shape[0], i_lengths.shape[0]])
lencl_ratios = np.zeros([inhibition_strengths.shape[0], i_lengths.shape[0]])
lenmb_ratios = np.zeros([inhibition_strengths.shape[0], i_lengths.shape[0]])

lenkernels_vdp = np.zeros([inhibition_strengths.shape[0], i_lengths.shape[0], stdp_delays.shape[0]])
lenkernels_cl = np.zeros([inhibition_strengths.shape[0], i_lengths.shape[0], stdp_delays.shape[0]])
lenkernels_mb = np.zeros([inhibition_strengths.shape[0],  i_lengths.shape[0], stdp_delays.shape[0]])


for i, wi in enumerate(inhibition_strengths): 
    for k , l_i in enumerate(i_lengths): 
        lenkernels_vdp[i,k], _ = inhib_STDP_sim_stim(stdp_delays, t_i,l_i, wi, e_strength,  E_L,'VDP')
        lenvdp_ratios[i,k] = LTP_LTD_ratio(lenkernels_vdp[i,k])
        lenkernels_cl[i,k], _ = inhib_STDP_sim_stim(stdp_delays, t_i,l_i, wi,e_strength, E_L,'Clopath')
        lencl_ratios[i,k] = LTP_LTD_ratio(lenkernels_cl[i,k])
        lenkernels_mb[i,k], _ = inhib_STDP_sim_stim(stdp_delays, t_i, l_i,wi,e_strength, E_L,'Meissner_Bernard')
        lenmb_ratios[i,k] = LTP_LTD_ratio(lenkernels_mb[i,k])

```
